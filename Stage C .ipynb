```python
df = df.drop(['stab'], axis=1)
```

```python
X = df.drop(['stabf'], axis = 1)
y = df['stabf']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train_df = encoder.fit_transform(y_train)
y_test_df = encoder.transform(y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_df = scaler.fit_transform(x_train)
train_df = pd.DataFrame(train_df, columns=x_train.columns)

x_test = x_test.reset_index(drop=True)
test_df = scaler.transform(x_test)
test_df = pd.DataFrame(test_df, columns=x_test.columns)
```

### Question 3

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=1)
regressor.fit(train_df, y_train_df)
y_pred = regressor.predict(test_df)
```

```python
y_pred
```

```python
y_pred_base = y_pred.round(0).astype('int')
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
accuracy = accuracy_score(y_true=y_test_df, y_pred=y_pred_base)
print('Accuracy using Random Forest regressor is: {}'.format(round(accuracy,4)))
```
